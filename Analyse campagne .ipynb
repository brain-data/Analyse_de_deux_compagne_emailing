### email sended 

```python
# importation de bibliothèque re  pour qu'on puisse utuliser Les expressions régulières
import re
```

```python
# la fréquence de chaque valeur dans la colonne ' Email sent '
df['Emails sent'].value_counts()
```

```python
# expressions régulière pour selectionner la chaine "Email #"  dans la colonne  Email sent
hashtag = re.compile(r"^(Email #)")
```

```python
# remplacer la phrase " not contacted yet "  par des zeros  
df['Emails sent']= df['Emails sent'].replace("not contacted yet","0") 
```

```python
#supprimer la chaine 'Email #' dans chauque ligne 

df['Emails sent'] = df['Emails sent'].apply(lambda x : re.sub(hashtag,"",str(x)))
```

```python
df['Emails sent']
```

```python
#remplacer les valeur null par -1
df['Emails sent'] = df['Emails sent'].apply(lambda x : x.replace('nan','-1'))
```

### Operation sur la colonne  Emails opened  #1

```python
# expressions régulière pour selectionner la chaine "Email #"  

hashtag2 = re.compile(r"(Email #)")
```

```python
# remplacer les valleur null par zero dans la colonne  Email opened
df['Emails opened'].fillna(0, inplace=True)
```

### Operation sur la colonne Opening count





```python
# remplacer les valeur null par zero
df['Opening count'].fillna(0, inplace=True)
```

```python
df['Opening count'].value_counts()
```

```python
df['Emails opened']
```

```python
# expressions régulière pour selectionner la chaine "Email #"  dans la colonne  Email opened
df['Emails opened'] = df['Emails opened'].apply(lambda x : re.sub(hashtag2,"",x) if x!=0 else x)
```

```python
# convertir  les colonne qui contient plusieur mail ouvert  en liste 
df['Emails opened']= df['Emails opened'].apply(lambda x : str(x).split(","))
```
